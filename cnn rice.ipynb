```python
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession

config=ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction=0.5
config.gpu_options.allow_growth=True
session= InteractiveSession(config=config)
```

```python
from tensorflow.keras.layers import Input,Lambda,Dense,Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
from glob import glob
```

```python
Image_size=[256,256]
train_path='conclusion neural network rice prediction/train'
test_path='conclusion neural network rice prediction/test'
```

```python
resnet=ResNet50(input_shape=Image_size+[3],weights='imagenet',include_top=False)
```

```python
for layer in resnet.layers:
    layer.trainabler=False
    
```

```python
folders=glob('conclusion neural network rice prediction/train')
```

```python
x=Flatten()(resnet.output)
```

```python
prediction=Dense(len(folders),activation='softmax')(x)
model=Model(inputs=resnet.input,outputs=prediction)
```

```python
len(folders)
```

```python
model.summary()
```

```python
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)    
```

```python
train_datagen=ImageDataGenerator(   #to read training image
        rescale=1/255,    #to get more generalized model so we are diving the given pixel by 255 in order to get the range from 0 to 1
        rotation_range=15,
        zoom_range=0.1,
        horizontal_flip=True
)

test_datagen=ImageDataGenerator(    #to read testing image
    rescale=1/255
)    
```

```python
 batch_size=32

train_set =train_datagen.flow_from_directory(
    'conclusion neural network rice prediction/train/',
     target_size=(256,256),
     batch_size=batch_size,
     class_mode='categorical'
)  
 
test_set =test_datagen.flow_from_directory(
    'conclusion neural network rice prediction/validation/',
     target_size=(256,256),
     batch_size=batch_size,
     class_mode='categorical'
)    
```
